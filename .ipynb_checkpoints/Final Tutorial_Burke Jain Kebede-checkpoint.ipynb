#### The 100-Point Scale 
According to the New York Times best-seller, "Wine Folly: The Essential Guide to Wine" by Madeline Puckette and Justin Hammack, The 100-point wine scale actually starts at 50 points. 
- *50-59:* wines are flawed and undrinkable
- *60-69:* wines are flawed and not recommended but drinkable
- *70-79:* wines are flawed and taste average
- *80-84:* wines are 'above average' to 'good'
- *90-94:* wines are 'superior' to 'exceptional' 
- *95-100:* wines are benchmark examples or 'classic'

Wine ratings do not necessarily indicate the deliciousness of a wine. Rather, they are scored based on production quality and typicity (i.e: how greatly the traits of a particular wine 'typify' the style and region is it from).

## i. Initial Installation of Libraries
To view our explanatory graphs and tables within our Jupyter Notebook, we must first import certain Python compatible libraries listed below. The documentation for said libraries can be found available online through a simple web search

```python
import pandas as pd
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
```

##  ii. Data Collection
### Loading Data
Once the neccessary library are properly imported, it is finally time to retrieve the data that that we will be getting from Kaggle. 

First, make sure that the winemag-data_first150k.csv file from Kaggle is downloaded and stored in the same directory. We will utilize pandas, which we imported above. Pandas is an open source, BSD-licensed library providing high-performance, easty-to-use data structures and data analysis tools for Python. In order to utilize the table of data, we will need to use the .read_csv function of pandas. 

Then, you must declare the names of columns.

For simplicity's sake, we took the first 10 rows of the wine review data to display here.

```python
#reading the csv file from local drive 
table = pandas.read_csv('winemag-data_first150k.csv')

#declaring names of the columns 
col = ['index','country', 'description', 'designation', 'points', 'price', 'province', 'region_1', 'region2', 'variety', 'winery']
table.columns = col

table.head(10)
```

## iii. Data Processing
### Data Tidying

We will need to tidy the data in order to remove extraneous data. Specifically, we remove null price values. We also got rid of regions, providences, the indexes, descriptions of the specific wines, as well as the winery. Again, for simplicity's sake, we only display the first ten rows of the data in this notebook.

```python
#We will now tidy up data

#Only keeping the columns that we will be using for our data analysis
table1 = table[['country', 'points', 'price', 'variety', 'winery']]

#Deleting rows where the value of price is null 
table1 = table1[pandas.notnull(table1['price'])]

table1.head(10)
```

## iv. Exploratory Data Analysis

### Price vs. Points
Time to explore our tidied data! Since the goal of this data exploration is to seek out what variable(s) create the highest point value, we will begin with the most intuitive research - price. Let's see if price really does make the wine...

Begin with the matplotlib library installed in the first part of this tutorial. We will choose our x and y values. For this plot, we chose x = price and y = points. We will mark these values with a period '.' We then use the .title, .xlabel, and .ylabel of the matplotlib in order to label our graph for ease of interpretation. 

Use plt.show() to show your new graph!

```python
#Plotting the Price vs. Points of the wines 

plt.plot(table1['price'],table1['points'], '.')

#Declaring the graph's title, x, and y labels
plt.title("Price vs. Points")
plt.xlabel('Price')
plt.ylabel('Points')
plt.show()

#Exploratory Data Analysis and Data Vis
#We need to determine if there is a relationship between the price of wine
#and the rating of the wine. 
```

Let's do another data visualization. We will import sklearn, or scikit-learn. We will utilize this tool in order to add a linear regression line to this graph. We will use the .fit function on the linear_model.LinearRegression() in order to create a plot with a reg line. We then use the .predict function to create a y prediction for each x value in price. We plot this value... Look below at our result... We can see that ouliers are really interfering with this display.

```python
import sklearn
from sklearn import linear_model

reg = linear_model.LinearRegression()

e2_X = [[x] for x in table1['price'].values]
e2_Y = [[y] for y in table1['points'].values]
regfit = reg.fit(e2_X, e2_Y)

y_pred = reg.predict(e2_X)

fig = plt.figure()
plt.plot(e2_X,e2_Y, '.')
plt.plot(e2_X, y_pred)
plt.xlabel("Price")
plt.ylabel("Points")
fig.tight_layout()
plt.title('Price vs Points')
plt.show()
```

The regression line above shows us that there is a linear trend. The cheaper the wine, especially the wines within the  range from $0 to $500, the worse the rating. We can see that as the wine becomes more and more expensive the better the ratings; However, after the price of $500 the ratings do not grow as quickly.  

```python
ggplot(aes(x='price', y='points'), data = table1) +\
    geom_violin() +\
    labs(title="Price vs Points",
         x = "price",
         y = "points")
```

Hm... These plots don't seem to suggest a linear relationship between points and price alone but let's make sure. We can import the class linear from Statsmodel. Statsmodel is a Python module that provides classes and functions for the estimation of statistical models, as well as for conductiing statistical tests, and statistical data exploration. That's what we're going to do here: statistical data exploration. 

We will use the Ordinary Least Squares (OLS) regression analysis. We will compare the correlation between points and price... Particularly, we will be taking a look at the R-Squared value of our formula. The R-squared value will give us some information about the goodness of fit of a model. The R2 coefficient of determination is a statistical measure of how well the regression line approximates the real data points. An R-squared of 1 indicates that the regression line perfectly fits the data. The R-squared value of this data is 0.211... This can be interpreted as "21% of the variation in the 'y' variable is explained by my model." This indicates that there is not a strong correlation between points and just price ... just as we thought. There must be something beyond JUST price into which we must look... let's take a look into countries!

```python
import statsmodels.formula.api as smf

est = smf.ols(formula = 'points ~ price', data = table1).fit()
est.summary()
```

### Country vs. Points
As we discussed earlier, wines are scored based on production quality and typicity. Typicity refers to how greatly the traits of a particular wine 'typify' the style and region is it from. Based on that information, lets see if the there is a correlation between certain countries, their prices, and the points. 

Let's first start with something simple: a bar chart of all the countries and the sums of all of the points per country. Let's see where the most highly-rated wines are coming from!
We first must manipulate the table... we groupby country, obviously.Next, we must sort by the point values in order to get the top values. Use matplotlib again. X-axis will be countries, Y-axis will be points. 

```python
table1 = table1.head(100)
table2 = table1.groupby(by=['country'], as_index=False)['points'].sum()
table2 = table2.sort_values(['points'], ascending=False)

table2.plot.bar(x = 'country', y = 'points',color='purple')
plt.xlabel("Countries")
plt.ylabel("Points")
plt.title('Points per country')
plt.show()
```

Interestingly, the US has the highest rated wines! Followed by Italy and France. Perhaps US millenials would enjoy this fact with pride. 

Next, we were curious about the distribution of points at each country. From the graph below we can see that the US although is leading in the best ratings there is a lot of variability,especially in comparison to other countries. The points for the US range from 86 to 96, on the other hand Argentina has a fraction of variability. 

```python
import ggplot
from ggplot import *

#box and whisker, meow
ggplot(aes(x='country', y='points'), data = table1.head(100)) +\
  geom_boxplot() +\
    labs(title="Country vs. Points",
         x = "Country",
         y = "Points")
```

After plotting the box and whiskers plot and seeing the variability amongst the countries we were curious about the the average
points per country. The graph below gives us some insight on the mean number of points per country. 

```python
table3 = table1.groupby(by=['country'], as_index=False)['points'].mean()
table3 = table3.sort_values(['points'], ascending=False)

items =['US', 'Italy', 'France', 'Spain', 'Portugal', 'Argentina']
#temp_table = table1.filter(items = ['country','points'] regex = items)

temp_table = table1[table1['country'].isin(items)]


table3 = temp_table.groupby(by=['country'], as_index=False)['points'].mean()
table3 = table3.sort_values(['points'], ascending=False)

table3.plot.bar(x = 'country', y = 'points',color='purple')
plt.xlabel("Countries")
plt.ylabel("Points")
plt.title('Points per country')
plt.show()
```

After seeing the average points per country we wanted to see what the average was for the variety of wines. So we decided to
plot that below and we got.... 

```python

table4 = table1.groupby(by=['variety'], as_index=False)['points'].mean()
table4 = table4.sort_values(['points'], ascending=False)

variety =['Pinot Noir', 'Cabernet Sauvignon', 'Chardonnay', 'Sangiovese', 'Tinta de Toro', 'Nebbiolo','Red Blend']
#temp_table = table1.filter(items = ['country','points'] regex = items)

temp_table_2 = table1[table1['variety'].isin(variety)]


table4 = temp_table_2.groupby(by=['variety'], as_index=False)['points'].mean()
table4 = table4.sort_values(['points'], ascending=False)

table4.plot.bar(x = 'variety', y = 'points',color='blue')
plt.xlabel("variety")
plt.ylabel("Points")
plt.title('Points per variety')
plt.show()
```

Next, we wanted to take an even closer look at this variation per country. We saw a similar trend as before where the better reviews occured at higher price points. We confirmed our earlier observation and realized that this trend is true for the 10 countries plotted. 

```python
countries =['US', 'Italy', 'France', 'Spain', 'Portugal']
table_temp = table1.copy()

for country in countries:
    temp_cont = table1.groupby(['country']).get_group(country)
   
    x_a = temp_cont['points'].values
    y_a = temp_cont['price'].values
    ln = np.polyfit(x = x_a, y = y_a, deg = 1)
    fn = np.poly1d(ln) 
    
    x_b = np.linspace(x_a.min(), x_a.max(), 100)
    y_b = fn(x_b)
    
    plt.figure(figsize=(10,5))
    plt.plot(x_a, y_a,'o',x_b,y_b)
    
    plt.xlabel("Points")
    plt.ylabel("Price")
    t = "Points vs. Price in %s" %(country)
    plt.title(t)
    plt.show()
```

Those plots look far more linearly related than those we created for price alone earlier. Let's use statsmodels again to see if there is a high enough R-Squared value....

Unfortunately, we still aren't there... The R-Squared value if 0.405. This can be interpreted as "41% of the variation in the 'y' variable is explained by my model," meaning that the regression lines we created did not fit the data in a good enough manner. The goodness of fit was subpar.

Hmm... we need to look into another factor with price other than country. 

```python
import statsmodels.formula.api as smf

est = smf.ols(formula = 'points ~ country*price', data = table1).fit()
est.summary()
```

### Variety of Wine vs. Points
Again, because wine is rated by typicity which is how greatly the traits of a particular wine 'typify' the style and region is it from... perhaps the VARIETY of the wine in combination with the price truly do lead to a correlation in higher point values. 

We approach this problem in a similar way as the countries. 
Let's begin with a bar plot... We see that the top-rated wines are Pinot Noir, Cabernet Sauvignon, and Chardonnay. 

```python
temp = table1.head(100)
table2 = temp.groupby(by=['variety'], as_index=False)['points'].sum()
table2 = table2.sort_values(['points'], ascending=False)

table2.plot.bar(x = 'variety', y = 'points',color='purple')
plt.xlabel("Variety")
plt.ylabel("Points")
plt.title('Total points per variety of wine')
plt.show()
```

Next, we will plot the linear regression lines for each variety of wine just as we did with the countries.

```python
variety =['Pinot Noir', 'Cabernet Sauvignon', 'Chardonnay', 'Sangiovese', 'Tinta de Toro', 'Nebbiolo', 'Red Blend' ]
table_temp = table1.copy()

for v in variety:
    temp_cont = table1.groupby(['variety']).get_group(v)
   
    x_a = temp_cont['points'].values
    y_a = temp_cont['price'].values
    ln = np.polyfit(x = x_a, y = y_a, deg = 1)
    fn = np.poly1d(ln) 
    
    x_b = np.linspace(x_a.min(), x_a.max(), 100)
    y_b = fn(x_b)
    
    plt.figure(figsize=(10,5))
    plt.plot(x_a, y_a,'o',x_b,y_b)
    
    plt.xlabel("Points")
    plt.ylabel("Price")
    t = "Points vs. Price for %s" %(v)
    plt.title(t)
    plt.show()
```

Wow, those look pretty good but again, let's check the R-Squared value.

At last! The R-Squared value ~ 0.8!! This can be interpreted as "80% of the variation in the 'y' variable is explained by my model." The goodness of fit is sufficient. There is a correlation between the top wines: Pinot Noir, Cabernet Sauvignon, Chardonnay, Sangiovese, Tinta de Toro, Nebbiolo, Red Blend and their high prices, and the points! The higher the price for these wines, the higher rated the wine will be. 

```python
est = smf.ols(formula = 'points ~ variety*price', data = table1).fit()
est.summary()
```
